```python
# Cambia el nombre del archivo PDF si es necesario
ruta_pdf = 'El Tribut - Fichas platos para Sala.pdf'
```

```python
import pdfplumber
import os
import google.generativeai as genai
from dotenv import load_dotenv
```

```python
load_dotenv()
api_key = os.getenv('GOOGLE_API_KEY')
```

```python
genai.configure(api_key=api_key)
```

```python
# Realizar una consulta simple
if 'api_key' in locals() and api_key:
    try:
        print("ü§ñ Haciendo consulta a Gemini (versi√≥n gratuita)...")
        
        # Crear el modelo (versi√≥n gratuita)
        model = genai.GenerativeModel('gemini-2.5-flash')
        
        # Hacer la consulta
        response = model.generate_content("Hola! Escribe un saludo amigable en espa√±ol")
        
        respuesta = response.text
        print("‚úÖ Respuesta recibida:")
        print("-" * 50)
        print(respuesta)
        print("-" * 50)
        
    except Exception as e:
        print(f"‚ùå Error en la consulta: {e}")
        print("üí° Si tienes error de cuota, aseg√∫rate de usar la versi√≥n gratuita del modelo")
else:
    print("‚ùå Primero debes configurar la API Key en la celda anterior")
```
